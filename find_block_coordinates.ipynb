```python
import pytesseract
import cv2
import numpy as np
from PIL import Image

def find_and_highlight_text_blocks(pdf_path, page_number, coords, final_tags):
    i = 0
    # Use Pytesseract to extract text and bounding box information
    image = cv2.imread(pdf_path)
    d = pytesseract.image_to_boxes(image)
   
    # Convert the image to a NumPy array
    image_array = np.array(image)
    
    # Iterate through the bounding box data and draw rectangles around text blocks
    for coord in coords:
        x, y, w, h = int(coord[0]), int(coord[1]), int(coord[2]), int(coord[3])
        cv2.rectangle(image_array, (x, y), (w, h), (0, 0, 255), 2)  # Draw red rectangles
        center_x = (x + w) // 2
        center_y = (y + h) // 2
        
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 0.5
        font_thickness = 1
        text_size = cv2.getTextSize(final_tags[i], font, font_scale, font_thickness)[0]
        text_x = center_x - text_size[0] // 2
        text_y = center_y + text_size[1] // 2
        
        cv2.putText(image_array, final_tags[i], (text_x, text_y), font, font_scale, (0, 125, 0), font_thickness, lineType=cv2.LINE_AA)
        i += 1

    # Save the modified image
    cv2.imwrite("highlighted_page.png", image_array)

# NO CLUE HOW TO MAKE THIS WORK FOR EVERYONE
pytesseract.pytesseract.tesseract_cmd = r'E:\programavimas\Tesseract OCR\tesseract.exe'
pdf_image_path = "sample3.jpg"
page_number = 1  # Adjust this to the desired page number

text = pytesseract.image_to_string(pdf_image_path)

print(text)

# Get bounding box estimates
#print(pytesseract.image_to_boxes(Image.open(pdf_image_path)))

# Get verbose data including boxes, confidences, line and page numbers
# print(pytesseract.image_to_data(Image.open(pdf_image_path)))

# Get information about orientation and script detection
# print(pytesseract.image_to_osd(Image.open(pdf_image_path)))

# WORKS SOMETHING
xml =pytesseract.image_to_pdf_or_hocr(pdf_image_path, extension='hocr')
print(xml)
#print(pytesseract.image_to_alto_xml(pdf_image_path))
```
